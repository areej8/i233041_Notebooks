# Task 2

```python
paths = {
    "W": {"distance": 300, "time": 25, "congestion_level": "Medium"},
    "X": {"distance": 200, "time": 15, "congestion_level": "Low"},
    "Y": {"distance": 150, "time": 20, "congestion_level": "Medium"},
    "Z": {"distance": 250, "time": 10, "congestion_level": "High"}
}
congestion_level = { "Low": 3, "Medium": 2, "High": 1 }

utility = lambda path: 100 - path["distance"] - path["time"] + (congestion_level[path["congestion_level"]] * 20)

scores = {}

for i, j in paths.items():
    scores[i] = utility(j)

best = max(scores, key=scores.get)

print("Utility Scores:")
for path, score in scores.items():
    print(f"Path {path}: {score}")

print("\nSelected Best Path:", best)
```

# Task 3

```python
plants = {
    "Section1": "Healthy",
    "Section2": "Wilted"
}
internal_state = {
    "Section1": "Unknown",
    "Section2": "Unknown"
}
sections = ["Section1", "Section2"]

for t in range(1, 4):
    print(f"Time Step {t}")
    for sec in sections:
        obs = plants[sec]
        internal_state[sec] = obs  
        
        if internal_state[sec] == "Wilted":
            action = "Water"
            plants[sec] = "Healthy"  
        else:
            action = "Move"
        
        print(f"{sec}: Observed = {obs}")
        print(f"Internal State = {internal_state[sec]}")
        print(f"Action = {action}")
```
